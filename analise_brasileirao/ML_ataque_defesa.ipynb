```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

filen_name = '../dados/static_serie_a_13_11_2023.csv'
dados = pd.read_csv(filen_name)
def analise_forca_ataque(dados, colunas_features, colunas_alvo):
    for coluna_alvo in colunas_alvo:
        # Seleção de Features para Força de Ataque
        X_ataque = dados[colunas_features]

        # Variável Alvo para Força de Ataque
        y_ataque = dados[coluna_alvo]

        # Divisão do Conjunto de Dados
        X_ataque_train, X_ataque_test, y_ataque_train, y_ataque_test = train_test_split(X_ataque, y_ataque, test_size=0.2, random_state=42)

        # Verifique se a coluna alvo está presente no DataFrame
        if coluna_alvo not in dados.columns:
            dados[coluna_alvo] = 0  # Ou utilize um valor padrão para preenchimento

        # Escolha do Modelo e Treinamento para Força de Ataque
        modelo_ataque = LinearRegression()
        modelo_ataque.fit(X_ataque_train, y_ataque_train)

        # Avaliação do Modelo para Força de Ataque
        y_ataque_pred = modelo_ataque.predict(X_ataque_test)
        erro_medio_quadratico_ataque = mean_squared_error(y_ataque_test, y_ataque_pred)
        print(f'Erro Médio Quadrático para Força de Ataque ({coluna_alvo}): {erro_medio_quadratico_ataque}')

        # Aplicação do Modelo para Criar a Coluna FA
        dados[f'FA_{coluna_alvo}'] = modelo_ataque.predict(dados[colunas_features])

    return dados

def analise_forca_defesa(dados, colunas_features, colunas_alvo):
    for coluna_alvo in colunas_alvo:
        # Seleção de Features para Força de Defesa
        X_defesa = dados[colunas_features]

        # Variável Alvo para Força de Defesa
        y_defesa = dados[coluna_alvo]

        # Divisão do Conjunto de Dados
        X_defesa_train, X_defesa_test, y_defesa_train, y_defesa_test = train_test_split(X_defesa, y_defesa, test_size=0.2, random_state=42)

        # Verifique se a coluna alvo está presente no DataFrame
        if coluna_alvo not in dados.columns:
            dados[coluna_alvo] = 0  # Ou utilize um valor padrão para preenchimento

        # Escolha do Modelo e Treinamento para Força de Defesa
        modelo_defesa = LinearRegression()
        modelo_defesa.fit(X_defesa_train, y_defesa_train)

        # Avaliação do Modelo para Força de Defesa
        y_defesa_pred = modelo_defesa.predict(X_defesa_test)
        erro_medio_quadratico_defesa = mean_squared_error(y_defesa_test, y_defesa_pred)
        print(f'Erro Médio Quadrático para Força de Defesa ({coluna_alvo}): {erro_medio_quadratico_defesa}')

        # Aplicação do Modelo para Criar a Coluna FD
        dados[f'FD_{coluna_alvo}'] = modelo_defesa.predict(dados[colunas_features])

    return dados

```

```python
dados['FTS'] = dados['FTS'].astype(str).str.rstrip('%').astype('float') / 100.0
dados['CS'] = dados['CS'].astype(str).str.rstrip('%').astype('float') / 100.0
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Suponha que você já tenha treinado o modelo e tenha os dados em um DataFrame chamado 'dados'
# Certifique-se de que 'dados' contém as colunas necessárias para as previsões (FA_W, FA_Pts, FA_L, FD_W, FD_Pts, FD_L)

def prever_partida(time_a, time_b, modelo_ataque, modelo_defesa, dados):
    # Encontrar as linhas correspondentes aos times
    dados_time_a = dados[dados.index == time_a]
    dados_time_b = dados[dados.index == time_b]

    # Extrair features para Força de Ataque
    features_ataque_time_a = dados_time_a[['FA_W', 'FA_Pts', 'FA_L']].values.reshape(1, -1)
    features_ataque_time_b = dados_time_b[['FA_W', 'FA_Pts', 'FA_L']].values.reshape(1, -1)

    # Extrair features para Força de Defesa
    features_defesa_time_a = dados_time_a[['FD_W', 'FD_Pts', 'FD_L']].values.reshape(1, -1)
    features_defesa_time_b = dados_time_b[['FD_W', 'FD_Pts', 'FD_L']].values.reshape(1, -1)

    # Previsão da Força de Ataque
    ataque_time_a = modelo_ataque.predict(features_ataque_time_a)
    ataque_time_b = modelo_ataque.predict(features_ataque_time_b)

    # Previsão da Força de Defesa
    defesa_time_a = modelo_defesa.predict(features_defesa_time_a)
    defesa_time_b = modelo_defesa.predict(features_defesa_time_b)

    # Calcular a probabilidade de vitória, empate e derrota
    prob_vitoria_time_a = ataque_time_a / defesa_time_b
    prob_vitoria_time_b = ataque_time_b / defesa_time_a
    prob_empate = 1 / (1 + (defesa_time_a / ataque_time_b) + (defesa_time_b / ataque_time_a))

    # Calcular a previsão da quantidade possível de gols
    gols_time_a = (ataque_time_a * defesa_time_b) / 2
    gols_time_b = (ataque_time_b * defesa_time_a) / 2

    # Imprimir os resultados
    print(f"Probabilidade de vitória do {time_a}: {prob_vitoria_time_a[0]:.2%}")
    print(f"Probabilidade de empate: {prob_empate[0]:.2%}")
    print(f"Probabilidade de vitória do {time_b}: {prob_vitoria_time_b[0]:.2%}")
    print(f"Previsão de gols: {gols_time_a[0]:.2f} - {gols_time_b[0]:.2f}")

# Exemplo de uso
# Suponha que 'modelo_ataque' e 'modelo_defesa' são os modelos treinados para Força de Ataque e Força de Defesa, respectivamente
time_a = 0  # Índice do time_a na tabela (por exemplo, Palmeiras)
time_b = 1  # Índice do time_b na tabela (por exemplo, Botafogo)

prever_partida(time_a, time_b, modelo_ataque, modelo_defesa, dados)
```
